```python
from bs4 import BeautifulSoup
import pandas 
import requests
url = 'https://www.espncricinfo.com/records/tournament/team-match-results/indian-premier-league-2023-15129'
page=requests.get(url)
soup=BeautifulSoup(page.text,'html.parser')
links=[]
table=soup.find_all('table')[0]
for row in table.find_all('tr')[1:]:
    href='https://www.espncricinfo.com'+row.find_all('td')[-1].find('a')['href']
    links.append(href)
# df = pd.DataFrame(MatchResults[1:], columns=MatchResults[0])
# df.to_csv('C:\Projects\web scrapping\Playerinfo.csv')
```

```python
def extract_batting_data(link,batting,l):
    s1=BeautifulSoup(requests.get(link).text,'html.parser')
    in_table=s1.find('table')
    
    for row in in_table.find_all('tr')[1:]:
        row_data=[]
        for col in row.find_all('td'):
            row_data.append(col.text.strip())
        if (l==len(row_data)):
            batting.append(row_data)
        else:
            continue
    return batting
s=BeautifulSoup(requests.get(links[1]).text,'html.parser')
table=s.find('table')
headings=[]
batting=[]
for row in table.find_all('th'):
    headings.append(row.text)
l=len(headings)
batting.append(headings)
# print(batting)
n=1
for link in links:
    # if (n<5):
    batting=extract_batting_data(link,batting,l)
        # n=n+1
    # else:
    # print(batting)
    # break
print(batting)
```
