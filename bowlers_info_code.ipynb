```python
from bs4 import BeautifulSoup
import requests
url = 'https://www.espncricinfo.com/records/tournament/team-match-results/indian-premier-league-2023-15129'
page=requests.get(url)
soup=BeautifulSoup(page.text,'html.parser')
links=[]
table=soup.find_all('table')[0]
for row in table.find_all('tr')[1:]:
    href='https://www.espncricinfo.com'+row.find_all('td')[-1].find('a')['href']
    links.append(href)
print(links)
```

```python
def extract_bowlers_data(link,bowling,l):
    s1=BeautifulSoup(requests.get(link).text,'html.parser')
    in_table=s1.find('table',class_="ds-w-full ds-table ds-table-md ds-table-auto")
    
    for row in in_table.find_all('tr')[1:]:
        row_data=[]
        for col in row.find_all('td'):
            row_data.append(col.text.strip())
        if (l==len(row_data)):
            bowling.append(row_data)
        else:
            continue
    return bowling
s=BeautifulSoup(requests.get(links[1]).text,'html.parser')
in_table2=s.find('table',class_="ds-w-full ds-table ds-table-md ds-table-auto")
# print(in_table2)
headings=[]
bowling=[]
for row in in_table2.find_all('th'):
    headings.append(row.text.strip())
# print(headings)
l=len(headings)
bowling.append(headings)
for link in links:
    bowling=extract_bowlers_data(link,bowling,l)
print(bowling)
```

```python
import pandas as pd
df=pd.DataFrame(bowling[1:],columns=bowling[0])
df.to_csv('C:\Projects\web scrapping\Bowling_info.csv')
```
